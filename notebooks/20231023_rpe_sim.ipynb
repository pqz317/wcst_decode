### Use some simulations to understand what a RPE-selective neuron really means

```python
import numpy as np
import matplotlib.pyplot as plt
```

RPE = R - Q

let R take on values either 0 or 1, Q take continuous values from 0.1 to 0.9

Plot "neuron" selectivity for all combinations of these values

```python
Q = np.arange(0, 1, 0.01)
R = np.array([0, 1])
```

```python
def normalize(nums):
    min = np.min(nums)
    max = np.max(nums)
    return (nums - min) / (max - min)
```

```python
Q_only = []
R_only = []
Q_plus_R = []
Q_minus_R = []
RPE = []
for q in Q: 
    for r in R:
        Q_only.append(q)
        R_only.append(r)
        Q_plus_R.append(q + r)
        Q_minus_R.append(q - r)
        RPE.append(r - q)

    
```

```python
fig, ax = plt.subplots(1, 4, figsize=(20, 5), sharex=False, sharey=False, tight_layout=True)
ax[0].scatter(RPE, normalize(Q_only), label= "Q only")
ax[0].legend()

ax[1].scatter(RPE, normalize(R_only), label="R only")
ax[1].legend()

ax[2].scatter(RPE, normalize(Q_plus_R), label="Q + R")
ax[2].legend()

ax[3].scatter(RPE, normalize(RPE), label="Q - R (RPE)")
ax[3].legend()

fig.supxlabel("RPE")
fig.supylabel("Normalized Resposne")
# ax.scatter(RPE, Q_plus_R, label="Q + R")
# ax.scatter(RPE, Q_minus_R, label="Q - R")
```
